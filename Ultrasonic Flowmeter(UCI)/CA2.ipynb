<h2>Meter A</h2>

<h6><b>MitoSheet</b></h6>

```python
A=df_a
#mitosheet.sheet(A)
#A.Shape
```

```python
A.rename(columns={"Flatness ratio": "Flatness_ratio"}, inplace=True)
A['AK'] = 0
A.rename(columns={"AK": "G_Diff_path_1"}, inplace=True)
A['G_Diff_path_1'] = A['G_St_path_1']-A['G_En_path_1']
A['AK'] = 0
A.rename(columns={"AK": "G_Diff_path_2"}, inplace=True)
A['G_Diff_path_2'] = A['G_St_path_2']-A['G_En_path_2']
A['AL'] = 0
A.rename(columns={"AL": "G_Diff_path_3"}, inplace=True)
A['G_Diff_path_3'] = A['G_St_path_3']-A['G_En_path_3']
A['AM'] = 0
A.rename(columns={"AM": "G_Diff_path_4"}, inplace=True)
A['G_Diff_path_4'] = A['G_St_path_4']-A['G_En_path_4']
A['BN'] = 0
A.rename(columns={"BN": "G_Diff_path_5"}, inplace=True)
A['G_Diff_path_5'] = A['G_En_path_5']-A['G_En_path_5']
A['BO'] = 0
A['G_Diff_path_5'] = A['G_St_path_5']-A['G_En_path_5']
A.rename(columns={"BO": "G_Diff_path_6"}, inplace=True)
A['G_Diff_path_6'] = A['G_St_path_6']-A['G_En_path_6']
A['BP'] = 0
A.rename(columns={"BP": "G_Diff_path_7"}, inplace=True)
A['G_Diff_path_7'] = A['G_St_path_7']-A['G_En_path_7']
A['BQ'] = 0
A.rename(columns={"BQ": "G_Diff_path_8"}, inplace=True)
A['G_Diff_path_8'] = A['G_St_path_8']-A['G_En_path_8']
A['AJ'] = 0
A.rename(columns={"AJ": "V_path_Avg"}, inplace=True)
A['V_path_Avg'] = (A['V_path_1']+A['V_path_2']+A['V_path_3']+A['V_path_4']+A['V_path_5']+A['V_path_6']+A['V_path_7']+A['V_path_8'])/8
A.drop(['S_path_1','S_path_2','S_path_3','S_path_4','S_path_5','S_path_6','S_path_7','S_path_8',
       'V_path_1','V_path_2','V_path_3','V_path_4','V_path_5','V_path_6','V_path_7','V_path_8',
       'G_St_path_1','G_St_path_2','G_St_path_3','G_St_path_4','G_St_path_5','G_St_path_6','G_St_path_7','G_St_path_8',
       'G_En_path_1','G_En_path_2','G_En_path_3','G_En_path_4','G_En_path_5','G_En_path_6','G_En_path_7','G_En_path_8',], axis=1, inplace=True)
  
```
